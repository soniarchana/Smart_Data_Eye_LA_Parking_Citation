```python
# Setting environment 
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import math
import datetime as dt
```

```python
# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# Loading csv datafile
parking_citations = pd.read_csv("../data/parking-citations.csv")
parking_citations.head(10)
```

```python
# Total records in file
print("Total records:", parking_citations.count() ) 
# 8777868 Rows
```

```python
# parking_citations.dtypes
parking_citations.info()
```

```python
# Renaming columns 
parking_citations=parking_citations.rename(index=str, \
    columns={"Ticket number": "Ticket_Number", "Issue Date": "Issue_Date", "Issue time": "Issue_Time",\
             "RP State Plate": "RP_State_Plate","Plate Expiry Date": "Plate_Expiry_Date", "Body Style": \
             "Body_Style", "Violation code":"Violation_Code", "Violation Description": "Violation_Desc", \
             "Fine amount": "Fine_Amount"})
parking_citations.head(5)
```

```python
# Dropping Columns that have a very high % of missing valuesnot needed - Marked Time, VIN, Meter Id
print('Number of columns before :{}'.format(len(parking_citations.columns)))
parking_citations.drop(['Meter Id','Marked Time','VIN'], axis=1, inplace=True)
print('Number of columns after :{}'.format(len(parking_citations.columns)))
parking_citations.head(5)
```

```python
# Which columns have NaN values
print("Columns which has NaN Values:", parking_citations.columns[parking_citations.isna().any()].tolist())
# Count of null / NaN values in columns
print("\nCount of NaN Values in columns:\n", parking_citations.isna().sum())
```

```python
# Removing rows with missing data
parking_citations = parking_citations.dropna(how="any")
parking_citations.count()   # 7897409 rows
```

```python
# Checking for duplicate ticket number
print("Number of tickets = ", parking_citations.Ticket_Number.count()) 
print("Number of unique tickets = ", parking_citations.Ticket_Number.nunique()) 
parking_citations.head(5)
```

```python
# Checking for Minimum and maximum Fine Amount

print("Minimum fine Amount: ", parking_citations["Fine_Amount"].min())
print("Maximum fine Amount: ", parking_citations["Fine_Amount"].max())
```

```python
# List Column Names
parking_citations.columns.values 
```

```python
# Create New Columns (dummies)
parking_citations["Issue_Date"] = pd.to_datetime(parking_citations["Issue_Date"])
print("Shape before:", parking_citations.shape)

parking_citations["Year_Issue_Date"] = parking_citations["Issue_Date"].map((lambda x: x.year))
parking_citations["Month_Issue_Date"] = parking_citations["Issue_Date"].map((lambda x: x.month))
parking_citations["Day_Issue_Date"] = parking_citations["Issue_Date"].map((lambda x: x.day))
parking_citations["DayOfWeek_Issue_Date"] = parking_citations["Issue_Date"].dt.dayofweek

print("Shape after:", parking_citations.shape)
parking_citations.head(5)
```

```python
# Converting new columns from float to int
parking_citations["Year_Issue_Date"] = parking_citations["Year_Issue_Date"].astype(np.int64)
parking_citations["Month_Issue_Date"] = parking_citations["Month_Issue_Date"].astype(np.int64)
parking_citations["Day_Issue_Date"] = parking_citations["Day_Issue_Date"].astype(np.int64)
parking_citations["DayOfWeek_Issue_Date"] = parking_citations["DayOfWeek_Issue_Date"].astype(np.int64)

parking_citations.info()
```

```python
parking_citations.head()
```

```python
# Checking Oldest and Latest Issue Date

print("Min Issue_Date:",min(parking_citations["Issue_Date"]))
print("Max Issue_Date:",max(parking_citations["Issue_Date"])) 
```

```python
# Checking Oldest and Latest Issue Year

print("Minimum Year: ", min(parking_citations["Year_Issue_Date"]))
print("Maximum Year: ", max(parking_citations["Year_Issue_Date"]))
      
parking_citations.Year_Issue_Date.unique()
```

```python
# Cleaning Latitude and Longitude by replacing 99999.000 to NaN

parking_citations['Latitude'] = np.where(parking_citations['Latitude']==99999.000, np.nan, parking_citations['Latitude'])
parking_citations['Longitude'] = np.where(parking_citations['Longitude']==99999.000, np.nan, parking_citations['Longitude'])
print("Lat/Long with 99999.000 cleaned")
```

```python
parking_citations.head(5)
```

```python
# Creating new column for latitude and longitude
# Transform coordinates from custom projection to lat long PyProj

import pyproj
# Coordinates in ft and false easting/northing in ft
x1,y1 = parking_citations['Latitude'].values, parking_citations['Longitude'].values
pm = '+proj=lcc +lat_1=34.03333333333333 +lat_2=35.46666666666667 +lat_0=33.5 +lon_0=-118 +x_0=2000000 +y_0=500000.0000000002 +ellps=GRS80 +datum=NAD83 +to_meter=0.3048006096012192 +no_defs'
x2,y2 = pyproj.transform(pyproj.Proj(pm,preserve_units = True), pyproj.Proj("+init=epsg:4326"), x1,y1)
parking_citations['NLatitude']=x2
parking_citations['NLongitude']=y2

print("Lan/long successfully transformed")
```

```python
# Checking which state plates have highest citations

df = parking_citations.groupby('RP_State_Plate')['RP_State_Plate'].count().sort_values(ascending=False).reset_index(name='count')
df.head()

# CA Plates have highest citation followed by NV, TX, AZ so it can be deduced that the drivers from these states
# are generally more prone to inviting parking tickets.
```
