```python
# Research

# Setting environment 
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
%matplotlib inline
import math
import datetime as dt
from collections import Counter
```

```python
# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# Loading csv datafile
df = pd.read_csv("../data/parking_citations_cleaned.csv")
df.head()
```

```python
# Dropping an column not needed 
print('Number of columns before:{}'.format(len(df.columns)))
df.drop('Unnamed: 0', axis=1, inplace=True)
print('Number of columns after :{}'.format(len(df.columns)))
```

```python
df.head()
```

```python
# Checking number of parking citations Issued each year 
# Picking top 5 years data for plotting
df_year = df.groupby("Year_Issue_Date")["Year_Issue_Date"].count().sort_values(ascending=False).nlargest(5).reset_index(name='count')
df_year

```

```python
# Plotting chart for number of parking citation yearwise for last 5 years (as have significant value)
plt.figure(figsize=(15,10))
sns.barplot(list(df_year["Year_Issue_Date"]), list(df_year["count"]))
plt.xlabel("Year")
plt.ylabel("Number Of Parking Tickets Given")
plt.title("Number Of Parking Tickets Issued Yearly")
plt.savefig("../imgs/PC_YearWise5yrs.png",bbox_inches='tight')
plt.show()
```

```python
# Checking number of parking citations Issued Month Wise in last 5 years

# # Month count for all years in dataset
# df_mon= df.groupby("Month_Issue_Date")["Month_Issue_Date"].count().sort_values(ascending=False).reset_index(name='count')
# df_mon

df_monthdata= df[(df['Year_Issue_Date'] >= 2015) & (df['Year_Issue_Date'] <= 2019)]
df_month = df_monthdata.groupby("Month_Issue_Date")["Month_Issue_Date"].count().sort_values(ascending=False).reset_index(name='count')
df_month
```

```python
# Plotting chart for number of parking citation monthwise in last 5 years (2015 to 2019)
plt.figure(figsize=(15,10))
sns.barplot(list(df_month["Month_Issue_Date"]), list(df_month["count"]))
plt.xlabel("Count of the parking citations")
plt.ylabel("Months")
plt.title("Parking citations issued monthwise consolidated")
plt.savefig("../imgs/PC_MonthWise.png",bbox_inches='tight')
plt.show()
```

```python
# Plotting the count of all state plates in specific year e.g. 2018

plate_count = Counter(df.loc[df["Year_Issue_Date"] == 2018]["RP_State_Plate"])
plt.figure(figsize=(10,10))
plt.pie(plate_count.values(), labels=plate_count.keys(), autopct='%1.2f%%')
fontP = FontProperties()
fontP.set_size('small')
plt.title("State plates for which the tickets were issued")
plt.savefig("../imgs/PC_StatePlateWise.png",bbox_inches='tight')
plt.show()
```

```python
# Plotting the count of State Plates other than CA for 2018
other_plate_count = Counter(df.loc[(df["Year_Issue_Date"] == 2018) &
                                                    (df["RP_State_Plate"] != 'CA')]["RP_State_Plate"])
# other_plate_count
plt.figure(figsize=(15,10))
sns.barplot(list(other_plate_count.keys()), list(other_plate_count.values()), color="skyblue")
plt.xticks(rotation=90)
plt.title("Parking Citations for State Plates other than CA")
plt.xlabel("States")
plt.ylabel("Count")
plt.savefig("../imgs/PC_OtherThanCAWise.png",bbox_inches='tight')
plt.show()
```

```python
# Count of parking citations on the Route 
df_route = df.groupby("Route")["Route"].count().sort_values(ascending=False).reset_index(name='count')
df_route.head(10)
```

```python
# Count of unique Location on each of the Route 
df_loc = df.groupby(["Route"])["Location"].nunique().sort_values(ascending=False).reset_index(name='count')
df_loc.head()
```

```python
# Count of parking citations at the top 10 Location in last 5 years

df_5yrsdata= df[(df['Year_Issue_Date'] >= 2015) & (df['Year_Issue_Date'] <= 2019)]
# df_5yrsdata
df_loc = df_5yrsdata.groupby(["Location"])["Location"].count().sort_values(ascending=False).nlargest(10).reset_index(name='count')
df_loc
```

```python
# Plotting chart for number of parking citations at the top 10 Location in last 5 years

plt.figure(figsize=(15,10))
sns.barplot(list(df_loc["Location"]), list(df_loc["count"]), order=df_loc["Location"])
plt.xticks(rotation=90)  
plt.xlabel("Location")
plt.ylabel("Number Of Parking Citations Given")
plt.title("Top 10 Locations and Number Of Parking Citations Issued (in last 5yrs)")
plt.savefig("../imgs/PC_LocationVsCitation.png",bbox_inches='tight')

plt.show()
```

```python
# Top 10 reason codes (Violation_Desc) for parking citation in the last month

# Pulling last months data in separate df
df_Lastyr = df[ (df['Year_Issue_Date'] == (max(df["Year_Issue_Date"])) ) ]
df_LastMon = df_Lastyr[ (df_Lastyr['Month_Issue_Date'] == (max(df_Lastyr["Month_Issue_Date"])) ) ]           
df_LastMon
```

```python
# Top 10 reason codes (Violation_Desc) for parking citation in the last month
df_ViolDesc = df_LastMon.groupby('Violation_Desc')['Ticket_Number'].count().sort_values(ascending = False).nlargest(10).reset_index(name='count')
df_ViolDesc

# whole dataset
# 0 NO PARK/STREET CLEAN	1894775
# 1	METER EXP.	1316444
# 2	PREFERENTIAL PARKING	508436
# 3	RED ZONE	479784
# 4	DISPLAY OF TABS	411449
```

```python
# Plotting top 10 reason codes (Violation_Desc) for parking citation in the last month

plt.figure(figsize=(15,10))
sns.barplot(list(df_ViolDesc["Violation_Desc"]), list(df_ViolDesc["count"]), color=(0.3, 0.3, 0.3, 0.3), order=df_ViolDesc["Violation_Desc"])
plt.xticks(rotation=90)  
plt.xlabel("Violation Description")
plt.ylabel("Number Of Parking Citations in Last Month")
plt.title("Parking Citations By Reason")
plt.savefig("../imgs/PC_VioDescWise.png",bbox_inches='tight')
plt.show()
```
